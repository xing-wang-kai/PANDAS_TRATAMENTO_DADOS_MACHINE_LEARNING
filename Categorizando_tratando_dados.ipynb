```python
columns_data_na = ["conta.contrato", "conta.faturamente_eletronico", "conta.metodo_pagamento"]
df_sem_nulls = data_not_empyt.dropna(subset=columns_data_na).copy()
df_sem_nulls.head()
```

```python
df_sem_nulls.reset_index(drop=True, inplace=True)
```

```python
df_sem_nulls
```

```python
df_sem_nulls.isna().sum()
```

```python
df_sem_nulls.describe()
```

```python
q1 = df_sem_nulls['cliente.tempo_servico'].quantile(0.25)
q3 = df_sem_nulls['cliente.tempo_servico'].quantile(0.75)
qlr = q3-q1
quantile_inferior = q1 - (1.5*qlr)
quantile_superior = q3 + (1.5*qlr)
```

```python
outliers_index = (df_sem_nulls['cliente.tempo_servico'] < quantile_inferior ) | ( df_sem_nulls['cliente.tempo_servico'] > quantile_superior )
outliers_index
```

```python
df_sem_nulls[outliers_index]
```

```python
df_sem_nulls[outliers_index][['cliente.tempo_servico']]
```

```python
df_sem_out = df_sem_nulls.copy()
```

```python
import numpy as np

df_sem_out.loc[outliers_index, 'cliente.tempo_servico'] = np.ceil(
    df_sem_out.loc[outliers_index, 'conta.cobranca.Total'] /
    df_sem_out.loc[outliers_index, 'conta.cobranca.mensal']
)

import seaborn as sns

sns.boxplot(x=df_sem_out['cliente.tempo_servico'])
```

```python
df_sem_out[outliers_index][['cliente.tempo_servico', 'conta.cobranca.Total', 'conta.cobranca.mensal']]
```

```python
q1 = df_sem_out['cliente.tempo_servico'].quantile(0.25)
q3 = df_sem_out['cliente.tempo_servico'].quantile(0.75)
qlr = q3-q1
quantile_inferior = q1 - (1.5*qlr)
quantile_superior = q3 + (1.5*qlr)

index_df_sem_out = (df_sem_out['cliente.tempo_servico'] < quantile_inferior ) | ( df_sem_out['cliente.tempo_servico'] > quantile_superior )
index_df_sem_out
```

```python
df_sem_out[index_df_sem_out][['cliente.tempo_servico', 'conta.cobranca.Total', 'conta.cobranca.mensal']]
```

```python
df_sem_out = df_sem_out[~index_df_sem_out]
df_sem_out.reset_index(drop=True, inplace=True)
```

```python
sns.boxplot(x=df_sem_out['cliente.tempo_servico'])
```

```python
df_sem_id = df_sem_out.drop('id_cliente', axis=1).copy()
```

```python
mapeamento = {
    'não':0,
    'sim': 1,
    "masculino":0,
    'feminino':1
}
```

```python
for cols in df_sem_id.columns:
  print(f"|--Coluna: {cols}")
  print(df_sem_id[cols].unique())
  print("|" + "-"*50 + "|")
```

```python
columns_binaria = ['Churn', "cliente.genero", "cliente.parceiro", "cliente.dependentes", "telefone.servico_telefone", "conta.faturamente_eletronico"]
```

```python
df_sem_id[columns_binaria] = df_sem_id[columns_binaria].replace(mapeamento)
```
